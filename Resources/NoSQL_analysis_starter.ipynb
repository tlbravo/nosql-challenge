### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": {"$eq":20}}
results = establishments.find(query)

# Use count_documents to display the number of documents in the result
document_number = establishments.count_documents(query)

# Display the first document in the results using pprint
for i in range(1):
    pprint(results[i])
```

```python
# Convert the result to a Pandas DataFrame
document_number_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print(f"Number of rows:{len(document_number_df)}")
# Display the first 10 rows of the DataFrame
document_number_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
london_query = {"LocalAuthorityName": {"$regex":"London"}, "RatingValue": {"$gte":4}}
london_results = establishments.find(london_query)

# Use count_documents to display the number of documents in the result
london_document_number = establishments.count_documents(london_query)
print(f"Number of documents:{london_document_number}")

# Display the first document in the results using pprint
pprint(establishments.find_one(london_query))
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.DataFrame(london_results)

# Display the number of rows in the DataFrame
print(f"Number of rows:{len(london_df)}")

# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# # Search within 0.01 degree on either side of the latitude and longitude. Latitude = 51.083812. Longitude = 1.195625.
# # Rating value must equal 5
# # Sort by hygiene score

degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

query = {"RatingValue" : "5", 
    "LocalAuthorityName": {"$lte": (latitude+degree_search), "$gte": (latitude-degree_search)},
    "geocode.longitude":{"$lte": (longitude+degree_search), "$gte": (longitude-degree_search)}
}

sort =  [{"scores.Hygiene" : 1}]
                                                                                     
restaurant_results = establishments.find(query).sort(sort).limit(5)                                                                                  

# Print the results
for restaurant in restaurant_results:
    print(restaurant)
```

```python
# Convert result to Pandas DataFrame
restaurant_hygiene_df = pd.DataFrame(restaurant_results)
```
